```python

import re

# Define your set of strings
set_of_strings = ["Harry Potter is a famous book", "The Harry Potter series is great", "I love reading Harry Potter"]

# Define the set of words you want to search for
words_to_match = ["Harry", "Potter", "is", "famous", "book"]

# Create a regular expression pattern to match any of the words
pattern = r'\b(?:' + '|'.join(map(re.escape, words_to_match)) + r')\b'

# Iterate through the set of strings and find matches
for input_string in set_of_strings:
    matches = re.findall(pattern, input_string)
    print(f"Matches in '{input_string}': {matches}")
```

```python
def parse_food_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    food_data = []
    food_item = {}

    for line in lines:
        line = line.strip()
        if line.startswith("No."):
            if food_item:
                food_data.append(food_item)
            food_item = {"No": line}
        elif line:
            key, value = line.split(':', 1)
            food_item[key.strip()] = value.strip()

    if food_item:
        food_data.append(food_item)

    return food_data

def find_matching_food_items(food_data, keyword):
    matching_items = []
    for item in food_data:
        if keyword.lower() in item['Name'].lower():
            matching_items.append(item)
    return matching_items

def convert_to_html_table(data):
    if not data:
        return "<p>No data to display</p>"

    # Start the HTML table
    html = "<table border='1'>\n"

    # Add the header row
    headers = data[0].keys()
    html += "<tr>\n"
    for header in headers:
        html += f"<th>{header}</th>\n"
    html += "</tr>\n"

    # Add the data rows
    for item in data:
        html += "<tr>\n"
        for value in item.values():
            html += f"<td>{value}</td>\n"
        html += "</tr>\n"

    # End the HTML table
    html += "</table>"

    return html
```

```python
#Use the function like this:
food_items = parse_food_data("category_jp_eng_cn.txt")
print(food_items)
```

```python
matches = find_matching_food_items(food_items, '親子丼')
print(matches)
```
